```python
import os
import numpy as np
from numpy.random import uniform, normal
import matplotlib.pyplot as plt
```

## Functions

```python
def generate_double_top(n_points):
    peak_1_width = int(n_points * uniform(0.15, 0.35))
    trough_1_width = int(n_points * uniform(0.40, 0.60))
    peak_2_width = int(n_points * uniform(0.65, 0.85))
    peak_1_height = uniform(2.95, 3.05)
    trough_1_height = uniform(0.95, 1.05)
    peak_2_height = uniform(2.95, 3.05)
    y = normal(0, uniform(0.15, 0.45), n_points)
    y[:peak_1_width] += np.linspace(0, peak_1_height, peak_1_width)  # ascent
    y[peak_1_width:trough_1_width] += np.linspace(peak_1_height, trough_1_height, trough_1_width - peak_1_width)  # descent
    y[trough_1_width:peak_2_width] += np.linspace(trough_1_height, peak_2_height, peak_2_width - trough_1_width)  # ascent
    y[peak_2_width:] += np.linspace(peak_2_height, peak_2_height - trough_1_height, n_points - peak_2_width)  # half-descent
    return y
```

```python
def generate_double_bottom(n_points):
    peak_1_width = int(n_points * uniform(0.15, 0.35))
    trough_1_width = int(n_points * uniform(0.40, 0.60))
    peak_2_width = int(n_points * uniform(0.65, 0.85))
    peak_1_height = uniform(2.95, 3.05)
    trough_1_height = uniform(0.95, 1.05)
    peak_2_height = uniform(2.95, 3.05)
    y = normal(0, uniform(0.15, 0.45), n_points)
    y[:peak_1_width] -= np.linspace(0, peak_1_height, peak_1_width)
    y[peak_1_width:trough_1_width] -= np.linspace(peak_1_height, trough_1_height, trough_1_width - peak_1_width)
    y[trough_1_width:peak_2_width] -= np.linspace(trough_1_height, peak_2_height, peak_2_width - trough_1_width)
    y[peak_2_width:] -= np.linspace(peak_2_height, peak_2_height - trough_1_height, n_points - peak_2_width)
    return y
```

```python
def generate_triple_top(n_points):
    peak_1_width = int(n_points * uniform(0.15, 0.25))
    trough_1_width = int(n_points * uniform(0.25 + 1e-10, 0.35))
    peak_2_width = int(n_points * uniform(0.40, 0.60))
    trough_2_width = int(n_points * uniform(0.65, 0.75))
    peak_3_width = int(n_points * uniform(0.75 + 1e-8, 0.85))
    peak_1_height = uniform(2.95, 3.05)
    trough_1_height = uniform(0.95, 1.05)
    peak_2_height = uniform(2.95, 3.05)
    trough_2_height = uniform(0.95, 1.05)
    peak_3_height = uniform(2.95, 3.05)
    y = normal(0, uniform(0.15, 0.45), n_points)
    y[:peak_1_width] += np.linspace(0, peak_1_height, peak_1_width)
    y[peak_1_width:trough_1_width] += np.linspace(peak_1_height, trough_1_height, trough_1_width - peak_1_width)
    y[trough_1_width:peak_2_width] += np.linspace(trough_1_height, peak_2_height, peak_2_width - trough_1_width)
    y[peak_2_width:trough_2_width] += np.linspace(peak_2_height, trough_2_height, trough_2_width - peak_2_width)
    y[trough_2_width:peak_3_width] += np.linspace(trough_2_height, peak_3_height, peak_3_width - trough_2_width)
    y[peak_3_width:] += np.linspace(peak_3_height, peak_3_height - trough_2_height, n_points - peak_3_width)
    return y
```

```python
def generate_triple_bottom(n_points):
    peak_1_width = int(n_points * uniform(0.15, 0.25))
    trough_1_width = int(n_points * uniform(0.25 + 1e-10, 0.35))
    peak_2_width = int(n_points * uniform(0.40, 0.60))
    trough_2_width = int(n_points * uniform(0.65, 0.75))
    peak_3_width = int(n_points * uniform(0.75 + 1e-8, 0.85))
    peak_1_height = uniform(2.95, 3.05)
    trough_1_height = uniform(0.95, 1.05)
    peak_2_height = uniform(2.95, 3.05)
    trough_2_height = uniform(0.95, 1.05)
    peak_3_height = uniform(2.95, 3.05)
    y = normal(0, uniform(0.15, 0.45), n_points)
    y[:peak_1_width] -= np.linspace(0, peak_1_height, peak_1_width)
    y[peak_1_width:trough_1_width] -= np.linspace(peak_1_height, trough_1_height, trough_1_width - peak_1_width)
    y[trough_1_width:peak_2_width] -= np.linspace(trough_1_height, peak_2_height, peak_2_width - trough_1_width)
    y[peak_2_width:trough_2_width] -= np.linspace(peak_2_height, trough_2_height, trough_2_width - peak_2_width)
    y[trough_2_width:peak_3_width] -= np.linspace(trough_2_height, peak_3_height, peak_3_width - trough_2_width)
    y[peak_3_width:] -= np.linspace(peak_3_height, peak_3_height - trough_2_height, n_points - peak_3_width)
    return y
```

```python
def generate_head_and_shoulders(n_points):
    peak_1_width = int(n_points * uniform(0.15, 0.25))
    trough_1_width = int(n_points * uniform(0.25 + 1e-10, 0.35))
    peak_2_width = int(n_points * uniform(0.40, 0.60))
    trough_2_width = int(n_points * uniform(0.65, 0.75))
    peak_3_width = int(n_points * uniform(0.75 + 1e-8, 0.85))
    peak_1_height = uniform(2.95, 3.05)
    trough_1_height = uniform(0.95, 1.05)
    peak_2_height = uniform(4.1, 4.4)
    trough_2_height = uniform(0.95, 1.05)
    peak_3_height = uniform(2.95, 3.05)
    y = np.random.normal(0, uniform(0.15, 0.45), n_points)
    y[:peak_1_width] += np.linspace(0, peak_1_height, peak_1_width)
    y[peak_1_width:trough_1_width] += np.linspace(peak_1_height, trough_1_height, trough_1_width - peak_1_width)
    y[trough_1_width:peak_2_width] += np.linspace(trough_1_height, peak_2_height, peak_2_width - trough_1_width)
    y[peak_2_width:trough_2_width] += np.linspace(peak_2_height, trough_2_height, trough_2_width - peak_2_width)
    y[trough_2_width:peak_3_width] += np.linspace(trough_2_height, peak_3_height, peak_3_width - trough_2_width)
    y[peak_3_width:] += np.linspace(peak_3_height, peak_3_height - trough_2_height, n_points - peak_3_width)
    return y
```

```python
def generate_inverse_head_and_shoulders(n_points):
    peak_1_width = int(n_points * uniform(0.15, 0.25))
    trough_1_width = int(n_points * uniform(0.25 + 1e-10, 0.35))
    peak_2_width = int(n_points * uniform(0.40, 0.60))
    trough_2_width = int(n_points * uniform(0.65, 0.75))
    peak_3_width = int(n_points * uniform(0.75 + 1e-8, 0.85))
    peak_1_height = uniform(2.95, 3.05)
    trough_1_height = uniform(0.95, 1.05)
    peak_2_height = uniform(4.1, 4.4)
    trough_2_height = uniform(0.95, 1.05)
    peak_3_height = uniform(2.95, 3.05)
    y = np.random.normal(0, 0.3, n_points)
    y[:peak_1_width] -= np.linspace(0, peak_1_height, peak_1_width)
    y[peak_1_width:trough_1_width] -= np.linspace(peak_1_height, trough_1_height, trough_1_width - peak_1_width)
    y[trough_1_width:peak_2_width] -= np.linspace(trough_1_height, peak_2_height, peak_2_width - trough_1_width)
    y[peak_2_width:trough_2_width] -= np.linspace(peak_2_height, trough_2_height, trough_2_width - peak_2_width)
    y[trough_2_width:peak_3_width] -= np.linspace(trough_2_height, peak_3_height, peak_3_width - trough_2_width)
    y[peak_3_width:] -= np.linspace(peak_3_height, peak_3_height - trough_2_height, n_points - peak_3_width)
    return y
```

```python
def generate_random(n_points):
    y = np.random.normal(0, 0.3, n_points)
    for i in range(1, n_points):
        trend = np.random.uniform(-0.002, 0.002)
        noise = np.random.uniform(-0.1, 0.1)
        y[i] = y[i - 1] + trend + noise
    return y
```

## Generate and Store Data and Images

```python
DATA_DIR = ".\\data\\"
IMG_DIR = ".\\images\\"
SUB_DIRS = ["double_top\\", "double_bottom\\", "triple_top\\", "triple_bottom\\", "head_and_shoulders\\", "inverse_head_and_shoulders\\", "random\\"]
NUM_ARRAYS = 5000
```

```python
for sub_dir in SUB_DIRS:
    data_path = os.path.join(DATA_DIR, sub_dir)
    if not os.path.exists(data_path):
        os.makedirs(data_path)
    img_path = os.path.join(IMG_DIR, sub_dir)
    if not os.path.exists(img_path):
        os.makedirs(img_path)
```

```python
for i in range(NUM_ARRAYS):
    n_points = np.random.randint(40, 201)
    arr = generate_double_top(n_points)
    arr_path = os.path.join(DATA_DIR, SUB_DIRS[0], str(i))
    np.save(arr_path, arr)
    img_path = os.path.join(IMG_DIR, SUB_DIRS[0], str(i))
    fig, ax = plt.subplots()
    ax.plot(arr)
    ax.axis('off')
    plt.savefig(img_path)
    plt.close()
```

## Visualize Patterns

```python
def parse_pattern(pattern):
    s = ''
    for word in pattern[:-1].split('_'):
        s += word + ' '
    return s[:-1]
```

```python
fig, ax = plt.subplots(5, 5, figsize=(32, 32))
for i in range(5):
    for j in range(5):
        pattern = np.random.choice(SUB_DIRS)
        index = np.random.randint(0, NUM_ARRAYS)
        ax[i][j].plot(np.load(DATA_DIR + pattern + f"{index}.npy"))
        ax[i][j].set_title(parse_pattern(pattern))
```
